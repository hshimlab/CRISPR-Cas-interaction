```python
import pandas as pd
from pandas import DataFrame
import seaborn as sns
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("1.Performance_evaluation_of_three_different_docking_tools.csv")


#Figure S8. Bar plot for comparing Fnat scores between tools.
custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(12,8)}
sns.set_theme(style="ticks", font_scale=1.3,rc=custom_params)

ax = sns.barplot(x="Structure", y="Fnat",
            hue="Tool", 
            data=df)
ax.set_facecolor('white')
plt.title('Bar plot for comparing Fnat scores between tools.\n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.show()

#Figure S8. Bar plot for comparing iRMSD scores between tools.
custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(12,8)}
sns.set_theme(style="ticks", font_scale=1.3,rc=custom_params)

ax = sns.barplot(x="Structure", y="iRMSD",
            hue="Tool",
            data=df)

plt.title('Bar plot for comparing iRMSD scores between tools.\n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.show()


#Figure S8. Bar plot for comparing LRMS scores between tools.
custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(12,8)}
sns.set_theme(style="ticks", font_scale=1.3,rc=custom_params)

ax = sns.barplot(x="Structure", y="LRMSD",
            hue="Tool", 
            data=df)

plt.title('Bar plot for comparing LRMS scores between tools.\n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.show()
```

```python

df = pd.read_csv("2.Performance_evaluation_of_16_tool_combinations.csv")


#Figure S9 iRMSD over Cas13a and combinations
df_irms = df.loc[df['Score_type'] == "iRMSD"]
df_irms = df_irms.loc[df['Cas_type'] == "Cas13a"]
df_irms['Structure'] = df_irms['Structure'].str.upper()

custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(14,9)}
sns.set_theme(style="ticks", font_scale=2,rc=custom_params)


ax = sns.barplot(y="RMSD", x="Structure",
            hue="Combination",
            order =["5WLH", "5W1I_AB", "5W1H", "5W1I_CD", "5WTK", "7OS0_CD", "6VRC", "6VRB", "5XWY", "7OS0_AF"],
            data=df_irms)
ax.set_facecolor('white')

plt.title('iRMSD over Cas13a and combinations\n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.show()


#Figure S10 iRMSD over Cas13b,d and combinations
df_lrms = df.loc[df['Score_type'] == "iRMSD"]
df_lrms = df_lrms.loc[df['Cas_type'] == "Cas13b,d"]
df_lrms['Structure'] = df_lrms['Structure'].str.upper()

custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(7,7)}
sns.set_theme(style="ticks", font_scale=1.5,rc=custom_params)

ax = sns.barplot(y="RMSD", x="Structure",
            hue="Combination",
            order =["6DTD","6IV9", "6IV8_AB", "6IV8_CD", "6E9E", "6AAY"],
            data=df_lrms)
ax.set_facecolor('white')
plt.title('iRMSD over Cas13b,d and combinations\n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.show()


#Figure S9 LRMSD over Cas13a and combinations
df_irms = df.loc[df['Score_type'] == "LRMSD"]
df_irms = df_irms.loc[df['Cas_type'] == "Cas13a"]
df_irms['Structure'] = df_irms['Structure'].str.upper()

custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(14,9)}
sns.set_theme(style="ticks", font_scale=2,rc=custom_params)

ax = sns.barplot(y="RMSD", x="Structure",
            hue="Combination",
            order =["5WLH", "5W1I_AB", "5W1H", "5W1I_CD", "5WTK", "7OS0_CD", "6VRC", "6VRB", "5XWY", "7OS0_AF"],
            data=df_irms)
ax.set_facecolor('white')
plt.title('LRMSD over Cas13a and combinations\n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.show()

#Figure S10 LRMSD over Cas13b,d and combinations
df_lrms = df.loc[df['Score_type'] == "LRMSD"]
df_lrms = df_lrms.loc[df['Cas_type'] == "Cas13b,d"]
df_lrms['Structure'] = df_lrms['Structure'].str.upper()


custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(7,7)}
sns.set_theme(style="ticks", font_scale=1.5,rc=custom_params)

ax = sns.barplot(y="RMSD", x="Structure",
            hue="Combination", 
            order =["6DTD","6IV9", "6IV8_AB", "6IV8_CD", "6E9E", "6AAY"],
            data=df_lrms)
ax.set_facecolor('white')

plt.title('LRMSD over Cas13b,d and combinations\n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)
plt.show()
```

```python
#Figure S11. Comparing averaged RMSD results (The less the better)
df = pd.read_csv("2.Performance_evaluation_of_16_tool_combinations.csv")

sns.set_theme(style="ticks", palette="pastel")

custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(12,8)}
sns.set_theme(style="ticks", font_scale=1.5,rc=custom_params)

ax = sns.barplot(y="Combination", x="RMSD",
            hue="Score_type",
            data=df)
ax.set_facecolor('white')
plt.title('Comparing averaged RMSD results (The less the better)\n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
#Figure S12-a. iRMSD over structures and template 

df = pd.read_csv("3.Evaluation_with_without_template_HDOCK.csv")

sns.set_theme(style="ticks", palette="pastel")
df_lrms = df.loc[df['Score_type'] == "iRMSD"]
df_lrms['Structure'] = df_lrms['Structure'].str.upper()

custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(7,7)}
sns.set_theme(style="ticks", font_scale=2,rc=custom_params)


ax = sns.barplot(y="RMSD", x="Structure",
            hue="Template", 
            data=df_lrms)
ax.set_facecolor('white')

plt.title('iRMSD over structures and template \n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)

plt.show()

#Figure S12-b. LRMSD over structures and template 

df_Lrms = df.loc[df['Score_type'] == "LRMSD"]
df_Lrms['Structure'] = df_Lrms['Structure'].str.upper()


custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(7,7)}
sns.set_theme(style="ticks", font_scale=2,rc=custom_params)

ax = sns.barplot(y="RMSD", x="Structure",
            hue="Template", 
            data=df_Lrms)
ax.set_facecolor('white')

plt.title('LRMSD over structures and template \n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.xticks(rotation=45)

plt.show()

#Figure S12-c.  

custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(12,8)}
sns.set_theme(style="ticks", font_scale=1,rc=custom_params)

ax = sns.catplot(y="Combination", x="RMSD",
            col='Score_type',
            row ='Template',
            kind="bar",
            data=df)

ax.fig.subplots_adjust(hspace=0.2, wspace=0.3)

```

```python
#Figure 4. [CA,P]iRMSD over Structures and Template

df = pd.read_csv("4.Performance_evaluation_of_docking_with_chosen_6_combinations.csv")

sns.set_theme(style="ticks", palette="pastel")

df_lrms = df.loc[df['Score_type'] == "iRMSD"]
df_lrms['Structure'] = df_lrms['Structure'].str.upper()

custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(14,7)}
sns.set_theme(style="ticks", font_scale=1.5,rc=custom_params)


ax = sns.boxplot(y="RMSD", x="Structure",
            hue="Template",
            order=["5W1H","5WLH","5W1I_AB","5W1I_CD"],
            data=df_lrms)

plt.title('[CA,P]iRMSD over structures and template \n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))
#plt.xticks(rotation=45)
plt.legend(frameon=False)

plt.show()

#[Not included]. [CA,P]LRMSD over Structures and Template

df_Lrms = df.loc[df['Score_type'] == "LRMSD"]
df_Lrms['Structure'] = df_Lrms['Structure'].str.upper()


custom_params = {"axes.spines.right": False, "axes.spines.top": False,'figure.figsize':(14,7)}
sns.set_theme(style="ticks", font_scale=1.5,rc=custom_params)


ax = sns.boxplot(y="RMSD", x="Structure",
            hue="Template",
            order=["5W1H","5WLH","5W1I_AB","5W1I_CD"],
            data=df_Lrms)

plt.title('[CA,P]LRMSD over structures and template \n',fontdict = {'fontsize': 20})
ax.legend(loc='best', bbox_to_anchor=(1, 0.5))

plt.legend(frameon=False)

plt.show()

```
